```python
import numpy as np
import pickle
from sklearn.cluster import DBSCAN
from sklearn.neighbors import LSHForest
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, fcluster
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
with open('state_changes.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
steps = [x[0] for x in data]
change_data = [x[1][0] for x in data]
```

```python
pca = PCA(n_components=2)
fit_data = pca.fit_transform(change_data)
x = fit_data[:,0]
y = fit_data[:,1]
```

# State Change Over Time

lighter = earlier time

darker = later time

```python
plt.figure(figsize=(16, 16))
sns.scatterplot(x, y, alpha=0.6, s=75, hue=steps, palette=sns.color_palette("hls", len(np.unique(steps))), legend=False)
```

```python
def dist(a, b):
    s = 0
    for i in range(len(a)):
        s += (a[i] - b[i]) ** 2
    return s ** 0.5
```

```python
dists = []
for i in range(2500):
    r1 = change_data[np.random.randint(len(change_data))]
    r2 = change_data[np.random.randint(len(change_data))]
    dists.append(dist(r1, r2))
```

```python
print(np.mean(dists), min(dists), max(dists), np.std(dists))
```

# State Change Linkage Clustering (dim = 2)

```python
z = linkage(fit_data[:10000], 'ward')
clusters = fcluster(z, 1, 'distance')

plt.figure(figsize=(14, 14))
sns.scatterplot(x[:10000], y[:10000], alpha=0.6, s=75, hue=clusters, palette=sns.color_palette("hls", max(clusters)), legend=False)
```

# State Change Linkage Clustering (dim = 128)

```python
z = linkage(change_data[:10000], 'ward')
clusters = fcluster(z, 1, 'distance')

plt.figure(figsize=(14, 14))
sns.scatterplot(x[:10000], y[:10000], alpha=0.6, s=75, hue=clusters, palette=sns.color_palette("hls", max(clusters)), legend=False)
```

# State Change DBScan Clustering (dim = 2)

```python
dbscan = DBSCAN(eps=0.005, min_samples = 2)
clusters = dbscan.fit_predict(fit_data[:10000])
print(max(clusters))

plt.figure(figsize=(14, 14))
sns.scatterplot(x[:10000], y[:10000], alpha=0.6, s=75, hue=clusters, palette=sns.color_palette("hls", len(np.unique(clusters))), legend=False)
```

# State Change DBScan Clustering (dim = 128)

```python
dbscan = DBSCAN(eps=0.001, min_samples = 2)
clusters = dbscan.fit_predict(change_data[:10000])
print(max(clusters))

plt.figure(figsize=(14, 14))
sns.scatterplot(x[:10000], y[:10000], alpha=0.6, s=75, hue=clusters, palette=sns.color_palette("hls", len(np.unique(clusters))), legend=False)
```

# State Change Sampling + LSHForest

```python
lshf = LSHForest(n_estimators=20, n_candidates=200,
                 n_neighbors=100).fit(change_data)
```

```python
clusters = [0 for _ in range(len(change_data))]

for i in range(1):
    center = change_data[np.random.randint(len(change_data))]
    neighbors = lshf.kneighbors(np.array(center).reshape(1, -1))[1]
    for n in neighbors:
        clusters[n] = i
```
