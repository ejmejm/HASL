```python
import numpy as np
import pickle
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
with open('state_changes.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
steps = [x[0] for x in data]
change_data = [x[1][0] for x in data]
```

```python
pca = PCA(n_components=2)
fit_data = pca.fit_transform(change_data)
x = fit_data[:,0]
y = fit_data[:,1]
```

```python
plt.figure(figsize=(16, 16))
sns.scatterplot(x, y, alpha=0.6, s=75, hue=steps)
```
